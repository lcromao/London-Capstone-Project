```python
import pandas as pd
```

```python
# Creating a correlation heatmap ploting the following datasets at the same canvas:
# health_df_2016, health_df_2017, health_df_2018, health_df_2019, health_df_2020, health_df_2021, health_df_2022

# Creating a correlation heatmap ploting the following datasets at the same canvas:


def correlation_heatmap(df):
    import seaborn as sns
    import matplotlib.pyplot as plt
    plt.figure(figsize=(8, 8))
    sns.heatmap(df.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.show()
```

### Análise do dataset de saúde

```python
health_df = pd.read_excel('data/WellBeing/Tabela Health.xlsx')
health_df = health_df.round(4)
```

```python
health_df_general = health_df[health_df['Borough'] == 'London'].drop(columns=['Borough', 'Year'])
health_df_general
```

```python
correlation_heatmap(health_df_general)
```

### Análise Financeira

```python
financial_df = pd.read_excel('data/WellBeing/Financial_Security.xlsx', sheet_name='Financial')
financial_df['WellBeing'] = health_df['Wellbeing']
financial_df = financial_df.round(4)
financial_df.drop(columns=['Rough Sleepers'], inplace=True)
```

```python
financial_df_general = financial_df[health_df['Borough'] == 'London'].drop(columns=['Borough', 'Year'])
financial_df_general
```

```python
correlation_heatmap(financial_df_general)
```
